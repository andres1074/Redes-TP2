```python
import numpy as np
import math as ma
import networkx as nx
import powerlaw
from matplotlib import pyplot as plt
import operator
import random
```

```python
# Files # Se abren los documentos
archive_bin="../redes2/dataset/yeast_Y2H_curado.txt"
archive_mul="../redes2/dataset/yeast_AP-MS_curado.txt"
archive_lit="../redes2/dataset/yeast_LIT_copy.txt"
archive_reg="../redes2/dataset/yeast_LIT_Reguly_curado.txt"
archive_ess="../redes2/dataset/Essential_ORFs_paperHe_curado.txt"
```

```python
# Funcion leer input files
def ldata(archive):
        f=open(archive)
	data=[]
	for line in f:
		line=line.strip()
		col=line.split()
		data.append(col)	
	return data
```

```python
###### Grafo bin
lista = ldata(archive_bin)
#print(lista)
Gbin = nx.Graph()
Gbin.add_edges_from(lista)


###### Grafo mul
lista = ldata(archive_mul)
#print(lista)
Gmul = nx.Graph()
Gmul.add_edges_from(lista)


###### Grafo lit
lista = ldata(archive_lit)
#print(lista)
Glit = nx.Graph()
Glit.add_edges_from(lista)


###### Grafo reg
lista = ldata(archive_reg)
#print(lista)
Greg = nx.Graph()
Greg.add_edges_from(lista)
```

```python
####Essential
###### Grafo
lista = ldata(archive_bin)
#print(lista)
listaess = ldata(archive_ess)
Gbin = nx.Graph()
Gbin.add_edges_from(lista)
#xcfbcbin=[]
#ycfbcbin=[]
N = Gbin.number_of_nodes()
for i,node in enumerate(listaess):
    if Gbin.has_node(node[0]):
        Gbin.remove_node(node[0])
    
    fracnodes= (i/float(N)) 
    Gcbin = max(nx.connected_component_subgraphs(Gbin), key=len) ###Giant component
    lcbin=len(Gcbin) ###Longitud de la componente gigante
    lccbin=lcbin/float(N)   #####Largest connected component
    xessbin=fracnodes
    yessbin=lccbin
    
 #   pltcfbcbin=plt.plot(x, y, 'k-', linewidth = 2)
```

```python
####degree centrality bin
###### Grafo 
lista = ldata(archive_bin)
#print(lista)
Gbin = nx.Graph()
Gbin.add_edges_from(lista)
dcbind=nx.degree_centrality(Gbin) ###Da la centralidad de los nodos pero con el mismo orden de los datos (d de desordenado)
dcbin = sorted(dcbind.items(), key=operator.itemgetter(1), reverse=True) ###Ordenar por mayor a menor centralidad
xdcbin=[]
ydcbin=[]
N = Gbin.number_of_nodes()
for i,node in enumerate(dcbin):
    Gbin.remove_node(node[0])
    fracnodes= (i/float(N)) 
    Gcbin = max(nx.connected_component_subgraphs(Gbin), key=len) ###Giant component
    lcbin=len(Gcbin) ###Longitud de la componente gigante
    lccbin=lcbin/float(N)   #####Largest connected component
    xdcbin.append(fracnodes)
    ydcbin.append(lccbin)
    
   # pltdcbin=plt.plot(x, y, 'k--', linewidth = 2)
```

```python
####eigenvector centrality bin
###### Grafo
lista = ldata(archive_bin)
#print(lista)
Gbin = nx.Graph()
Gbin.add_edges_from(lista)
ecbind=nx.eigenvector_centrality(Gbin)
ecbin = sorted(ecbind.items(), key=operator.itemgetter(1), reverse=True) ###Ordenar por mayor a menor centralidad
xecbin=[]
yecbin=[]
N = Gbin.number_of_nodes()
for i,node in enumerate(ecbin):
    Gbin.remove_node(node[0])
    fracnodes= (i/float(N)) 
    Gcbin = max(nx.connected_component_subgraphs(Gbin), key=len) ###Giant component
    lcbin=len(Gcbin) ###Longitud de la componente gigante
    lccbin=lcbin/float(N)   #####Largest connected component
    xecbin.append(fracnodes)
    yecbin.append(lccbin)
    
#    pltecbin=plt.plot(x, y, 'k-', linewidth = 2)
```
